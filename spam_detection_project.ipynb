```python
def get_data(path):
    data = []
    files = os.listdir(path)
    for file in files:
        with open(path+file, encoding = "ISO-8859-1") as f:
          mail = f.read()
          data.append(mail)
    return data
```

<h1 style="color:black;font-size:45px;font-family:Georgia;text-align:center;"><strong>3.<strong style="color:#19e6e6;font-size:40px;font-family:Georgia;">Clean <strong style="color:black;font-size:45px;font-family:Georgia;">The <strong style="color:#19e6e6;font-size:40px;font-family:Georgia;">Dataset <strong style="color:black;font-size:45px;font-family:Georgia;"> </strong></strong></strong></strong></strong></h1>

```python
class MailCleaner(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.stemmer = PorterStemmer()
        self.lemmatizer = WordNetLemmatizer()
    def fit(self, X, y=None): 
        return self
    def transform(self, X):
        text_list = []
        for mail in X:
            b = email.message_from_string(mail)
            body = ""

            if b.is_multipart():
                for part in b.walk():
                    ctype = part.get_content_type()
                    cdispo = str(part.get('Content-Disposition'))

                    # skip any text/plain (txt) attachments
                    if ctype == 'text/plain' and 'attachment' not in cdispo:
                        body = part.get_payload(decode=True)  # get body of email
                        break
            else:
                body = b.get_payload(decode=True) # get body of email
            
            soup = BeautifulSoup(body, "html.parser") #supprimer les balises HTML
            text = soup.get_text()
            text =text.lower() # lowerCase 
            text = text.replace('$', " dollar ") # remplacer le $ par 'dollar'
            text = re.sub(r'(https|http)?:\/\/(\w|\.|\/|\?|\=|\&|\%)*\b', " httpaddr ", text, flags=re.MULTILINE) #remplace les URL par 'httpaddr'
            text = re.sub(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', ' emailaddr ', text, flags=re.MULTILINE) #remplac les e-mails par 'emailaddr'
            text = re.sub(r'\b[0-9]\b', ' nombre ', text, flags=re.MULTILINE) #remplace nombres par 'nombre'
            text = text.translate(str.maketrans('', '', string.punctuation)) # eliminer les espaces et ponctuation
            text = ''.join([i for i in text if not i.isdigit()]) # remove digits
            stop_words = stopwords.words('english')
            words_list = [w for w in text.split() if w not in stop_words] # supprimer stop words : and , the ...
            words_list = [self.lemmatizer.lemmatize(w) for w in words_list] #lemmatization
            words_list = [self.stemmer.stem(w) for w in words_list] #Stemming
            text_list.append(words_list)
        return text_list
```

```python
mail_cleaner = MailCleaner()
text_ham = mail_cleaner.transform(ham)
text_spam = mail_cleaner.transform(spam)
text_easy_ham = mail_cleaner.transform(easy_ham)
text_hard_ham = mail_cleaner.transform(hard_ham)
```

<h1 style="color:black;font-size:25px;font-family:Georgia;text-align:center;"><strong>4.1 Define a dictionary</strong></h1>

```python
def define_a_dict(sources,k=2):
    all_words =[]
    for source in sources:
        for word_list in source:
            all_words.extend(word_list)
    
    bag_of_words = Counter(all_words)
    vocab = [word for word,v in bag_of_words.items() if v>k ]
    word_to_index =dict()
    # mapper chaque mot dans l’email prétraité à son index dans une liste d'index de mots 
    for idx , word in enumerate(vocab):
        word_to_index[word]=idx
    return word_to_index
```

```python
def percent_value_counts(df, feature):
    """This function takes in a dataframe and a column and finds the percentage of the value_counts"""
    percent = pd.DataFrame(round(df.loc[:,feature].value_counts(dropna=False, normalize=True)*100,2))
    ## creating a df with th
    total = pd.DataFrame(df.loc[:,feature].value_counts(dropna=False))
    ## concating percent and total dataframe

    total.columns = ["Total"]
    percent.columns = ['Percent']
    return pd.concat([total, percent], axis = 1)
```

<p style= "font-family:Georgia;color:#000000;font-size:120%;text-align:center;"><b>Q1 - Does the GRE score affect the probability of getting Admission? (chi square test)</b><br></p>
    
    
    
<p style= "font-family:Georgia;color:#000000;font-size:110%;text-align:center;border-radius:10px 10px;border-style: dotted;border-width:5px;border-color:#19e6e6;">
    1.H0: GREScore and Admission_categories are independent.<br>
2.H1: GREScore and Admission_categories dependent.<br>
 </p>

```python
fig = px.histogram(df, x="GREScore", color ="Admission_categories",barmode='group',color_discrete_map = {"High":'#0000ff',"Medium":'#0099ff',"Low": "#4dffff"})

fig.update_layout(bargap = 0.5,title='GREScore Vs Admission_categories',
                  template='plotly_white')
fig.show()
```

```python
# Create a table of cross-tabulations
Admission_Gre = pd.crosstab(df["Admission_categories"],df["GREScore"])
print(Admission_Gre)
```

```python
stat, p, dof, expected = chi2_contingency(Admission_Gre)
print('Degress of freedom = %d' % dof)
print('Expected frequencies ', expected)
```

```python
prob = 0.95
critical = chi2.ppf(0.95, dof)
print('probability=%.3f, critical=%.3f, stat=%.3f' % (prob, critical, stat))
```

```python
# interpret test-statistic
if abs(stat) >= critical:
    print('The variables are dependent (We reject H0)')
else:
    print('The variables are independent (We fail to reject H0)')
```

```python
# interpret p-value
alpha = 1.0 - prob
print('Significance = %.3f, p = %.3f' % (alpha, p))
if p <= alpha:
    print('The variables are dependent (We reject H0)')
else:
    print('The variables are independent (We fail to reject H0)')
```

<p style= "font-family:Georgia;color:#000000;font-size:110%;text-align:center;"><b>We can see that,</b><br>
-The P-value is lower than the significance level of 5%, so we can reject the null hypothesis with a confidence of 95%.<br>
-This means, that GREScore and getting Admission are dependent with a confidence level of 95%.
 </p>

<h1 style="color:black;font-size:45px;font-family:Georgia;text-align:center;"><strong>14.<strong style="color:#19e6e6;font-size:40px;font-family:Georgia;">Split <strong style="color:black;font-size:45px;font-family:Georgia;">The <strong style="color:#19e6e6;font-size:40px;font-family:Georgia;">Dataset <strong style="color:black;font-size:45px;font-family:Georgia;">Into <strong style="color:#19e6e6;font-size:40px;font-family:Georgia;">Train-Test <strong style="color:black;font-size:45px;font-family:Georgia;">:-</strong></strong></strong></strong></strong></strong></strong></h1>

```python
# Assigning the featurs as X and trarget as y
X= df.drop(["ChanceofAdmit"],axis =1)
y= df["ChanceofAdmit"]
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25, random_state=7)
```

<h1 style="color:black;font-size:45px;font-family:Georgia;text-align:center;"><strong>15.<strong style="color:#19e6e6;font-size:40px;font-family:Georgia;">Model <strong style="color:black;font-size:45px;font-family:Georgia;">Creation <strong style="color:#19e6e6;font-size:40px;font-family:Georgia;">:- </strong></strong></strong></strong></h1>

```python
#Building piplines for model selection

pipeline_lr=Pipeline([("scalar1",StandardScaler()),
                      ("LR",LinearRegression())])

pipeline_dt=Pipeline([("scalar2",StandardScaler()),
                      ("DT",DecisionTreeRegressor())])

pipeline_rf=Pipeline([("scalar3",StandardScaler()),
                      ("RF",RandomForestRegressor())])

pipeline_knn=Pipeline([("scalar4",StandardScaler()),
                       ("KN",KNeighborsRegressor())])

pipeline_xgb=Pipeline([("scalar5",StandardScaler()),
                       ("XGB",XGBRegressor())])

#List of all the pipelines
pipelines = [pipeline_lr, pipeline_dt, pipeline_rf, pipeline_knn, pipeline_xgb]

# Dictionary of pipelines and model types for ease of reference
pipe_dict = {0: "LinearRegression", 1: "DecisionTree", 2: "RandomForest",3: "KNeighbors", 4: "XGBRegressor"}


# Fit the pipelines
for pipe in pipelines:
    pipe.fit(X_train, y_train)

#Getting CV scores    
cv_results_rms = []
for i, model in enumerate(pipelines):
    cv_score = cross_val_score(model, X_train,y_train,scoring="neg_root_mean_squared_error", cv=10)
    cv_results_rms.append(cv_score)
    print("%s: %f " % (pipe_dict[i], cv_score.mean()))
```


<p style= "font-family:Georgia;color:#000000;font-size:120%;text-align:center;">The best model is ... Linear Regression 🏆
 </p>

<h1 style="color:black;font-size:45px;font-family:Georgia;text-align:center;"><strong>16.<strong style="color:#19e6e6;font-size:40px;font-family:Georgia;">Evaluate <strong style="color:black;font-size:45px;font-family:Georgia;">The <strong style="color:#19e6e6;font-size:40px;font-family:Georgia;">Model <strong style="color:black;font-size:45px;font-family:Georgia;">:-</strong></strong></strong></strong></strong></h1>

```python
# Model prediction on test data
pred = pipeline_lr.predict(X_test)
# Model Evaluation
r2 = metrics.r2_score(y_test, pred)
Adjusted_r2 = 1 - (1-metrics.r2_score(y_test, pred))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
mae = metrics.mean_absolute_error(y_test, pred)
mse = metrics.mean_squared_error(y_test, pred)
rmse = np.sqrt(metrics.mean_squared_error(y_test, pred))

# initialise data of lists.
ResultData = [[r2],[Adjusted_r2],[mae],[mse],[rmse]]
# Creates pandas DataFrame.
Results = pd.DataFrame(ResultData,columns= ["Scores"] ,index = ["R-Squared","Adjusted R-Squared", "Mean Absolute Error","Mean Square Error","Root Mean Square Error"])
Results.style.background_gradient(cmap="Greys")
```
